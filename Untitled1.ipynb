```python
# 8. Ordene o DataFrame por alguma coluna

import pandas as pd
data = {
    'Nome': ['Mall', 'Mel', 'Gab', 'Fab', 'AA'],
    'Idade': [18, 17, 16, 19, 18],
    'Numero': [25, 28, 9, 30, 23]}
df = pd.DataFrame(data)
df.sort_values('Idade')
```

```python
# 9. Agrupe os dados por alguma coluna e calcule a m√©dia

import pandas as pd
data = {
    'Nome': ['Mall', 'Mel', 'Gab', 'Fab', 'AA'],
    'Idade': [18, 17, 16, 19, 18],
    'Numero': [25, 28, 9, 30, 23]}
df = pd.DataFrame(data)
df.groupby('Numero').mean('Numero')
```
